```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.histplot(df['TotalSpend'],bins=20,kde=True,color='darkblue')
plt.title("Distribution of TotalSpend")
plt.xlabel("TotalSpend")
plt.ylabel("Frequency")
```

```python
sns.countplot(x='Membership',data=df, palette='Set2')
plt.title("Membership")
plt.xlabel("Membership")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(x='Gender',y='TotalSpend',data=df, palette='Set2')
plt.title("TotalSpend by Gender")
plt.xlabel("Gender")
plt.ylabel("TotalSpend")
plt.show()
```

```python
Q1=df['TotalSpend'].quantile(0.25)
Q2=df['TotalSpend'].median()
Q3=df['TotalSpend'].quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
outliers=df[(df['TotalSpend']<lower_bound) | (df['TotalSpend']>upper_bound)]
print("Lower Bound:",lower_bound)
print("Upper Bound:",upper_bound)
print("Outliers:")
print(outliers)
outliers_removed = df[(df['TotalSpend'] >= lower_bound) & (df['TotalSpend'] <= upper_bound)]
plt.figure(figsize=(8, 5))
sns.histplot(outliers_removed['TotalSpend'], bins=15, kde=True, color='darkblue')
plt.axvline(lower_bound, color='red', linestyle='--', label='Lower Bound')
plt.axvline(upper_bound, color='green', linestyle='--', label='Upper Bound')
plt.title('TotalSpend Distribution (Outliers Removed)')
plt.xlabel('TotalSpend')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
